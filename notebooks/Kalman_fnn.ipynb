```python
# FNN model for kalman filtering
import numpy as np
import tensorflow as tf

#The Kalman filter parameters
A = 0.9  # State transition matrix
H = 1.0  # Observation matrix
Q = 0.01  # Process noise covariance
R = 1.0  # Measurement noise covariance

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

#model.save('./kalman_fnn.h5')

def kalman_filter(x_est, P_est, z):
    x_pred = A * x_est
    P_pred = A * P_est * A + Q

    K = P_pred * H / (H * P_pred * H + R)
    x_est = x_pred + K * (z - H * x_pred)
    P_est = (1 - K * H) * P_pred

    return x_est, P_est

# Simulate data
true_value = np.sin(np.linspace(0, 10, 100))
observations = true_value + np.random.normal(0, np.sqrt(R), 100)


x_est = 0.0
P_est = 1.0

# Training loop
for z in observations:
    x_est, P_est = kalman_filter(x_est, P_est, z)

    x_pred_nn = model.predict(np.array([x_est]))[0, 0]

    model.train_on_batch(np.array([x_est]), np.array([z]))

    print(f"True Value: {true_value} | Kalman Filter Estimate: {x_est} | NN Estimate: {x_pred_nn}")

```
